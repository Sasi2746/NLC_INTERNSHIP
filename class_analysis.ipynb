# Class Performance Analysis

This notebook creates random class marks data, saves it to Excel, reads it back, and performs a simple basic data analysis project with graphs.

## 1. Import Libraries and Set Up Data

We start by importing the libraries we need for analysis and graphing.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
np.random.seed(42)
```

## 2. Generate Random Class Score Data

Here we create a synthetic class dataset with marks in five subjects and a final grade.

```python
student_names = [
    'Aarav', 'Aisha', 'Arjun', 'Diya', 'Ishaan', 'Kavya', 'Meera', 'Rohan',
    'Saanvi', 'Vihaan', 'Anika', 'Kabir', 'Nisha', 'Rahul', 'Priya', 'Dev',
    'Ira', 'Siddharth', 'Neha', 'Krish', 'Ananya', 'Yash', 'Pooja', 'Aman',
    'Tanya', 'Mohan', 'Sneha', 'Ved', 'Riya', 'Harsh'
]

classes = ['10-A', '10-B', '10-C']
subjects = ['English', 'Math', 'Science', 'Social', 'Computer']

records = []
for idx, name in enumerate(student_names, start=1):
    marks = np.random.randint(55, 101, size=len(subjects))
    total = marks.sum()
    average = round(total / len(subjects), 2)
    percentage = round((total / 500) * 100, 2)

    if percentage >= 90:
        grade = 'A+'
    elif percentage >= 80:
        grade = 'A'
    elif percentage >= 70:
        grade = 'B'
    elif percentage >= 60:
        grade = 'C'
    else:
        grade = 'D'

    record = {
        'StudentID': f'ST{idx:03d}',
        'StudentName': name,
        'Class': np.random.choice(classes),
        'English': marks[0],
        'Math': marks[1],
        'Science': marks[2],
        'Social': marks[3],
        'Computer': marks[4],
        'Total': total,
        'Average': average,
        'Percentage': percentage,
        'Grade': grade,
    }
    records.append(record)

class_df = pd.DataFrame(records)
class_df.head()
```

## 3. Save Synthetic Data to Excel and Load It

This makes the notebook feel like a real Excel-based class project.

```python
excel_file = 'class_scores_random.xlsx'
class_df.to_excel(excel_file, index=False)

loaded_df = pd.read_excel(excel_file)
loaded_df.head()
```

## 4. Inspect and Clean the Dataset

We check the structure of the data and confirm it is ready for analysis.

```python
loaded_df.shape, loaded_df.columns.tolist(), loaded_df.isnull().sum().sum(), loaded_df.dtypes
```

## 5. Compute Class Analytics and Top 5 Scorers

Now we calculate summary statistics and find the top 5 students by total score.

```python
summary_stats = loaded_df[['English', 'Math', 'Science', 'Social', 'Computer', 'Total', 'Average', 'Percentage']].agg(['mean', 'median', 'min', 'max'])
top_5_scorers = loaded_df.sort_values(by='Total', ascending=False).head(5)
subject_means = loaded_df[['English', 'Math', 'Science', 'Social', 'Computer']].mean().sort_values(ascending=False)

top_5_scorers[['StudentID', 'StudentName', 'Class', 'Total', 'Average', 'Percentage', 'Grade']], summary_stats, subject_means
```

## 6. Visualize Results with Graphs

We use simple charts to make the analysis easy to understand.

```python
fig, axes = plt.subplots(1, 3, figsize=(21, 6))

# Bar chart of top 5 scorers
sns.barplot(data=top_5_scorers, x='StudentName', y='Total', palette='viridis', ax=axes[0])
axes[0].set_title('Top 5 Scorers')
axes[0].set_xlabel('Student Name')
axes[0].set_ylabel('Total Marks')
axes[0].tick_params(axis='x', rotation=45)

# Histogram of total marks
sns.histplot(loaded_df['Total'], bins=8, kde=True, color='steelblue', ax=axes[1])
axes[1].set_title('Distribution of Total Scores')
axes[1].set_xlabel('Total Marks')
axes[1].set_ylabel('Number of Students')

# Subject-wise average marks
subject_means.plot(kind='bar', color='coral', ax=axes[2])
axes[2].set_title('Average Marks by Subject')
axes[2].set_xlabel('Subject')
axes[2].set_ylabel('Average Marks')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

plt.figure(figsize=(7, 7))
grade_counts = loaded_df['Grade'].value_counts().sort_index()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Grade Distribution')
plt.show()
```
